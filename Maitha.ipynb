```python
# Importing necessary libraries
from datetime import datetime
from colorama import Fore, Back, Style
```

```python
# Passenger class to represent passengers
class Passenger:
    def __init__(self, name, ticketNumber, seatNumber):
        self.name = name
        self.ticketNumber = ticketNumber
        self.seatNumber = seatNumber

    def getName(self):          # Method to get passenger's name
        return self.name

    def getTicketNumber(self):  # Method to get passenger's ticket number
        return self.ticketNumber

    def getSeatNumber(self):    # Method to get passenger's seat number
        return self.seatNumber

    def setName(self, name):    # Method to set passenger's name
        self.name = name

    def setTicketNumber(self, ticketNumber):  # Method to set passenger's ticket number
        self.ticketNumber = ticketNumber

    def setSeatNumber(self, seatNumber):      # Method to set passenger's seat number
        self.seatNumber = seatNumber

# Airport class to represent airports
class Airport:
    def __init__(self, name):
        self.name = name
        self.flights = []

    def getName(self):      # Method to get airport's name
        return self.name

    def getFlights(self):   # Method to get list of flights
        return self.flights

# Flight class to represent flights
class Flight:
    def __init__(self, flightNumber, departureAirport, destinationAirport, departureDateTime, arrivalDateTime, gate):
        self.flightNumber = flightNumber
        self.departureAirport = departureAirport
        self.destinationAirport = destinationAirport
        self.departureDateTime = departureDateTime
        self.arrivalDateTime = arrivalDateTime
        self.gate = gate

    def getFlightNumber(self):          # Method to get flight's number
        return self.flightNumber

    def getDepartureAirport(self):      # Method to get departure airport
        return self.departureAirport

    def getDestinationAirport(self):    # Method to get destination airport
        return self.destinationAirport

    def getDepartureDateTime(self):     # Method to get departure date and time
        return self.departureDateTime

    def getArrivalDateTime(self):       # Method to get arrival date and time
        return self.arrivalDateTime

    def getGate(self):                   # Method to get gate number
        return self.gate

    def setFlightNumber(self, flightNumber):    # Method to set flight's number
        self.flightNumber = flightNumber

    def setDepartureAirport(self, departureAirport):  # Method to set departure airport
        self.departureAirport = departureAirport

    def setDestinationAirport(self, destinationAirport):  # Method to set destination airport
        self.destinationAirport = destinationAirport

    def setDepartureDateTime(self, departureDateTime):    # Method to set departure date and time
        self.departureDateTime = departureDateTime

    def setArrivalDateTime(self, arrivalDateTime):        # Method to set arrival date and time
        self.arrivalDateTime = arrivalDateTime

    def setGate(self, gate):                              # Method to set gate number
        self.gate = gate

# BoardingPassSystem class to represent boarding pass system
class BoardingPassSystem:
    def __init__(self, passengerName, flightNumber, departureAirport, destinationAirport, arrivalDateTime, gate, seatNumber, electronicTicket):
        self.passengerName = passengerName
        self.flightNumber = flightNumber
        self.departureAirport = departureAirport
        self.destinationAirport = destinationAirport
        self.arrivalDateTime = arrivalDateTime
        self.gate = gate
        self.seatNumber = seatNumber
        self.electronicTicket = electronicTicket

    def getPassengerName(self):         # Method to get passenger's name
        return self.passengerName

    def getFlightNumber(self):          # Method to get flight's number
        return self.flightNumber

    def getDepartureAirport(self):      # Method to get departure airport
        return self.departureAirport

    def getDestinationAirport(self):    # Method to get destination airport
        return self.destinationAirport

    def getArrivalDateTime(self):       # Method to get arrival date and time
        return self.arrivalDateTime

    def getGate(self):                   # Method to get gate number
        return self.gate

    def getSeatNumber(self):            # Method to get seat number
        return self.seatNumber

    def getElectronicTicket(self):      # Method to get electronic ticket
        return self.electronicTicket

    def setPassengerName(self, passengerName):   # Method to set passenger's name
        self.passengerName = passengerName

    def setFlightNumber(self, flightNumber):     # Method to set flight's number
        self.flightNumber = flightNumber

    def setDepartureAirport(self, departureAirport):  # Method to set departure airport
        self.departureAirport = departureAirport

    def setDestinationAirport(self, destinationAirport):  # Method to set destination airport
        self.destinationAirport = destinationAirport

    def setArrivalDateTime(self, arrivalDateTime):    # Method to set arrival date and time
        self.arrivalDateTime = arrivalDateTime

    def setGate(self, gate):                          # Method to set gate number
        self.gate = gate

    def setSeatNumber(self, seatNumber):              # Method to set seat number
        self.seatNumber = seatNumber

    def setElectronicTicket(self, electronicTicket):  # Method to set electronic ticket
        self.electronicTicket = electronicTicket
```

```python
# Complete Passenger Details
passenger_name = input("Enter passenger name: ")        # Prompt user to input passenger name
ticket_number = input("Enter ticket number: ")           # Prompt user to input ticket number
seat_number = input("Enter seat number: ")               # Prompt user to input seat number
passenger = Passenger(passenger_name, ticket_number, seat_number)  # Create passenger object with provided details
print("\t\t Complete Passenger Details")       # Print message for test case
print(f"{Fore.YELLOW}Name:{Style.RESET_ALL}", passenger.getName())  # Print passenger name
print(f"{Fore.YELLOW}Ticket Number:{Style.RESET_ALL}", passenger.getTicketNumber())  # Print ticket number
print(f"{Fore.YELLOW}Seat Number:{Style.RESET_ALL}", passenger.getSeatNumber())      # Print seat number
```

```python
# Test Case 2: Invalid Departure Date and Time
flight_number = input("\nEnter flight number: ")                         # Prompt user to input flight number
departure_airport = input("Enter departure airport: ")                   # Prompt user to input departure airport
destination_airport = input("Enter destination airport: ")               # Prompt user to input destination airport
departure_date_time = input("Enter departure date and time (YYYY-MM-DD HH:MM): ")  # Prompt user to input departure date and time
arrival_date_time = input("Enter arrival date and time (YYYY-MM-DD HH:MM): ")      # Prompt user to input arrival date and time
gate = input("Enter gate: ")                                             # Prompt user to input gate
try:
    departure_date_time = datetime.strptime(departure_date_time, '%Y-%m-%d %H:%M')   # Convert departure date and time to datetime object
    arrival_date_time = datetime.strptime(arrival_date_time, '%Y-%m-%d %H:%M')         # Convert arrival date and time to datetime object
    flight = Flight(flight_number, departure_airport, destination_airport, departure_date_time, arrival_date_time, gate)  # Create flight object with provided details
    print("\nTest Case 2: Invalid Departure Date and Time")           # Print message for test case
    print(f"{Fore.YELLOW}Flight Number:{Style.RESET_ALL}", flight.getFlightNumber())   # Print flight number
    print(f"{Fore.YELLOW}Departure Airport:{Style.RESET_ALL}", flight.getDepartureAirport())  # Print departure airport
    print(f"{Fore.YELLOW}Destination Airport:{Style.RESET_ALL}", flight.getDestinationAirport())  # Print destination airport
    print(f"{Fore.YELLOW}Departure Date and Time:{Style.RESET_ALL}", flight.getDepartureDateTime())  # Print departure date and time
    print(f"{Fore.YELLOW}Arrival Date and Time:{Style.RESET_ALL}", flight.getArrivalDateTime())  # Print arrival date and time
    print(f"{Fore.YELLOW}Gate:{Style.RESET_ALL}", flight.getGate())  # Print gate
except ValueError:
    print("Invalid date or time format. Please enter in YYYY-MM-DD HH:MM format.")  # Handle invalid date or time format
```

```python
# Complete Boarding Pass Information
boarding_pass = BoardingPassSystem(passenger.getName(), flight.getFlightNumber(), flight.getDepartureAirport(), 
                                    flight.getDestinationAirport(), flight.getArrivalDateTime(), flight.getGate(), 
                                    passenger.getSeatNumber(), passenger.getTicketNumber())

print("\n" + Back.BLUE + Fore.WHITE + "\tThe detail of the Boarding Pass" + Style.RESET_ALL)  # Print boarding pass detail header
print(f"{Fore.YELLOW}Passenger:{Style.RESET_ALL}", boarding_pass.getPassengerName())  # Print passenger name
print(f"{Fore.YELLOW}Flight:{Style.RESET_ALL}", boarding_pass.getFlightNumber())  # Print flight number
print(f"{Fore.YELLOW}Departure Airport:{Style.RESET_ALL}", boarding_pass.getDepartureAirport())  # Print departure airport
print(f"{Fore.YELLOW}Destination Airport:{Style.RESET_ALL}", boarding_pass.getDestinationAirport())  # Print destination airport
print(f"{Fore.YELLOW}Arrival Time:{Style.RESET_ALL}", boarding_pass.getArrivalDateTime())  # Print arrival time
print(f"{Fore.YELLOW}Gate:{Style.RESET_ALL}", boarding_pass.getGate())  # Print gate
print(f"{Fore.YELLOW}Seat Number:{Style.RESET_ALL}", boarding_pass.getSeatNumber())  # Print seat number
print(f"{Fore.YELLOW}Electronic Ticket:{Style.RESET_ALL}", boarding_pass.getElectronicTicket())  # Print electronic ticket
```
